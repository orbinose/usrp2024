```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#We load the data
dataold = np.genfromtxt("/home/jamesdr/usrp-sciprog/data.txt")
```

```python
#We remove the column titles
data = np.delete(dataold, 0,0)
```

```python
#We load the variables
x = data[:,0]
y = data[:,1]
y_std = data[:,2]
```

```python
#We plot
plt.scatter(x, y, color='red')
plt.errorbar(x, y, y_std, solid_capstyle='projecting', color = 'red', capsize=5, linestyle='none')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sample Data")
plt.grid()
```

```python
#We find the parabola of best fit
p = np.polyfit(x, y, 2)
```

```python
def line(x):
    poly = p[0]*x**2 + p[1]*x + p[2]
    return poly
```

```python
#We sort x for neatness with our polynomial
x_new = np.sort(x)

#We plot
plt.scatter(x, y, color='red')
plt.errorbar(x, y, y_std, solid_capstyle='projecting', color = 'red', capsize=5, linestyle='none')
plt.plot(x_new, line(x_new), color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sample Data")
plt.grid()
```
